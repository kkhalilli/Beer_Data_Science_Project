Question 1
Rank the top 3 breweries which produce the strongest beers (a beer's strength is determined by its ABV content).

```python
df_q1 = df.groupby(by=["beer_brewerId"], as_index = False)["beer_ABV"].mean().sort_values(by="beer_ABV", ascending=False)
df_q1 = df_q1.rename(columns = {"beer_ABV": "mean_brewer_beerABV"})
df_q1.nlargest(3,"mean_brewer_beerABV")["beer_brewerId"]
```

```python
# plot the distribution of the beer ABV that different breweries are making
_ = df_q1.boxplot(column = "mean_brewer_beerABV", figsize=(10, 7))
```

Question 2
Which year did beers enjoy the highest ratings?

```python
df["review_year"] = pd.to_datetime(df["review_time"], unit="s").dt.year
```

```python
reviews_per_year = df.groupby("review_year").size().to_frame("reviews_per_year").reset_index()
reviews_per_year
```

```python
df_q2 = df.groupby("review_year")["review_overall"].mean().reset_index().sort_values(by = "review_overall", ascending = False)
df_q2 = df_q2.rename(columns = {"review_overall": "yearly_mean_overall_review"}) 
df_q2
```

```python
df_q2 = df_q2.merge(reviews_per_year, how = 'inner')
total_count = df_q2['reviews_per_year'].sum()
df_q2["weighted_average"] = df_q2['reviews_per_year'] / total_count
df_q2["weighted_rating"] = df_q2["yearly_mean_overall_review"] * df_q2["weighted_average"]
best_years = df_q2.nlargest(3,"weighted_rating")['review_year']
best_years
```

```python
from matplotlib import pyplot as plt
figure = plt.figure(figsize = (13,8))
ax1 = figure.add_subplot(111)
ax2 = ax1.twinx()
df_q2.plot(x = "review_year", y = "yearly_mean_overall_review", kind = "bar", color = "blue", ax=ax1, width = 0.25, position = 0)
df_q2.plot(x="review_year", y="reviews_per_year", kind="bar", color="red", ax=ax2, width=0.25, position=1)
```

Question 3
Based on the users' ratings, which factors are important among taste, aroma, appearance, and palette?

```python
# there are multiple reviews per beer so we aggregate the reviews
df_q3 = df.groupby(by="beer_beerId")[[
    "review_taste",
    "review_aroma",
    "review_appearance",
    "review_palette",
    "review_overall"
]].mean()

df_q3.corr()
```

```python
from seaborn import heatmap
_ = heatmap(data = df_q3.corr(), annot = True)
```

Question 4
If you were to recommend 3 beers to your friends based on this data, which ones would you recommend?

```python
df_q4 = df.groupby('beer_beerId').size().to_frame().rename(columns = {0:'count'})
rating = df.groupby('beer_beerId')['review_overall'].mean().reset_index()
df_q4 = df_q4[df_q4['count'] >  10]
df_q4 = df_q4.reset_index()
df_q4 = df_q4.merge(rating, how = 'inner')
df_q4.describe()
df_q4 = df_q4.sort_values(['review_overall','count'], ascending = False)
df_q4
```
